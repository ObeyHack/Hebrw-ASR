# Plot

```python
import pandas as pd
def plot_loss(df, y_max, title):
    # plot 2 columns for training and validation loss
    import matplotlib.pyplot as plt
    plt.plot(df['Train'], label='training')
    plt.plot(df['Val'], label='validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')

    # set y-axis limits
    plt.ylim([0, y_max])
    
    plt.legend()

    # set title
    plt.title(title)
    
    plt.show()
```

```python
def plot_decoding(greedy, beam, beam_lm):
    import matplotlib.pyplot as plt
    import numpy as np

    labels = ['CER', 'WER']
    bars = {
    'Greedy': greedy,
    'Beam': beam,
    'Beam with LM': beam_lm
    }

    x = np.arange(len(labels))  # the laael locations
    width = 0.25  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    for attribute, measurement in bars.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, measurement, width, label=attribute)
        ax.bar_label(rects, padding=3)
        multiplier += 1

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_title('Decoding Performance')
    ax.set_xticks(x + width, labels)

    # tight layout
    plt.tight_layout()



    ax.legend()
    plt.show()
```

# Simple NN - large scale

```python
simple_nn_df = pd.read_csv('Simple_NN.csv')
plot_loss(simple_nn_df, 6, 'Simple NN Loss')
```

# Simple NN - small scale

```python
simple_small_nn_df = pd.read_csv('simple_small.csv')
plot_loss(simple_small_nn_df, 20, 'Simple Small NN')
```

# DeepSpeech baseline

```python
deep_nn_df = pd.read_csv('deep_nn.csv')
plot_loss(deep_nn_df, 'Deep NN')
```

# Optimizer tuning

```python
deep_tuned_nn_df = pd.read_csv('deep_tuned_nn.csv')
plot_loss(deep_tuned_nn_df, 'Deep Tuned NN')
```

# Decoding

```python
# [CER, WER]
greedy = [0.5, 0.6]
beam = [0.3, 0.4]
beam_lm = [0.2, 0.3]
plot_decoding(greedy, beam, beam_lm)
```
